```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Step 1: Fetch Bitcoin price data
def fetch_data():
    df = yf.download("BTC-USD", period="1y", interval="1d")
    df = df[['Close']].dropna()
    df.columns = ['close']
    return df

# Step 2: Preprocess data
def preprocess_data(df, sequence_length=60):
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(df)

    X, y = [], []
    for i in range(sequence_length, len(scaled_data)):
        X.append(scaled_data[i-sequence_length:i])
        y.append(scaled_data[i])

    X, y = np.array(X), np.array(y)
    return X, y, scaler

# Step 3: Build LSTM model
def build_model(input_shape):
    model = Sequential()
    model.add(LSTM(units=50, return_sequences=True, input_shape=input_shape))
    model.add(LSTM(units=50))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

# Step 4: Predict next N days
def predict_next_days(model, last_sequence, scaler, n_days=15):
    predictions = []
    curr_seq = last_sequence.copy()

    for _ in range(n_days):
        pred_scaled = model.predict(curr_seq.reshape(1, curr_seq.shape[0], 1), verbose=0)
        pred_price = scaler.inverse_transform(pred_scaled)[0][0]
        predictions.append(pred_price)

        # Update sequence
        pred_scaled_for_next = pred_scaled.reshape(1)
        curr_seq = np.append(curr_seq[1:], pred_scaled_for_next)
        curr_seq = curr_seq.reshape(-1, 1)

    return predictions

# Step 5: Run it all
def main():
    df = fetch_data()
    X, y, scaler = preprocess_data(df)

    # Split data (80% train, 20% test)
    split = int(0.8 * len(X))
    X_train, X_test = X[:split], X[split:]
    y_train, y_test = y[:split], y[split:]

    model = build_model((X.shape[1], 1))
    model.fit(X_train, y_train, epochs=5, batch_size=32)

    # Predict and evaluate
    y_pred = model.predict(X_test)
    y_pred_inv = scaler.inverse_transform(y_pred)
    y_test_inv = scaler.inverse_transform(y_test)

    # Plot predictions vs actual
    plt.figure(figsize=(10, 5))
    plt.plot(y_test_inv, label='Actual Price')
    plt.plot(y_pred_inv, label='Predicted Price')
    plt.title("Bitcoin Price Prediction")
    plt.xlabel("Time Steps")
    plt.ylabel("Price (USD)")
    plt.legend()
    plt.show()

    # Predict next 15 days
    last_sequence = X[-1]
    future_predictions = predict_next_days(model, last_sequence, scaler, n_days=15)

    print("\nNext 15 Days Bitcoin Price Predictions (Daily Close):")
    for i, price in enumerate(future_predictions, 1):
        print(f"Day {i}: ${price:.2f}")

    # Plot future predictions
    plt.figure(figsize=(10, 5))
    plt.plot(range(1, 16), future_predictions, marker='o')
    plt.title("Next 15 Days Bitcoin Price Predictions")
    plt.xlabel("Days Ahead")
    plt.ylabel("Predicted Close Price (USD)")
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    main()
```

```python
import numpy as np
import pandas as pd
import yfinance as yf
import time
import os
import csv
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.preprocessing import MinMaxScaler
from datetime import datetime

# Function to fetch data using yfinance
def fetch_data(ticker="BTC-USD", period="7d", interval="1m"):
    try:
        # Fetch data from yfinance
        data = yf.download(ticker, period=period, interval=interval)
        return data
    except Exception as e:
        print(f"ERROR: Failed to fetch data - {e}")
        return None

# Preprocess the data (scaling)
def preprocess_data(df):
    # Scale the 'Close' price
    scaler = MinMaxScaler(feature_range=(0, 1))
    data_scaled = scaler.fit_transform(df[['Close']].values)

    # Prepare data for LSTM
    def create_dataset(data, time_step=60):
        X, y = [], []
        for i in range(len(data) - time_step - 1):
            X.append(data[i:(i + time_step), 0])
            y.append(data[i + time_step, 0])
        return np.array(X), np.array(y)

    X, y = create_dataset(data_scaled)
    X = X.reshape(X.shape[0], X.shape[1], 1)

    return X, y, scaler

# Build LSTM model
def build_model(input_shape):
    model = Sequential()
    model.add(LSTM(units=50, return_sequences=True, input_shape=input_shape))
    model.add(LSTM(units=50, return_sequences=False))
    model.add(Dense(units=1))
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

# Save prediction to CSV file
def save_prediction_to_csv(y_test_inv, y_pred_inv):
    actual_price = y_test_inv[-1][0]  # Last actual price in test set
    predicted_price = y_pred_inv[-1][0]  # Last predicted price
    timestamp = datetime.now().strftime('%Y-%m-%d %H:%M:%S')

    filename = 'btc_price_predictions.csv'
    header = ['Timestamp', 'Actual Price', 'Predicted Price']

    # Create file with header if it doesn't exist
    try:
        with open(filename, 'x', newline='') as file:
            writer = csv.writer(file)
            writer.writerow(header)
    except FileExistsError:
        pass  # Already exists, skip creating header

    # Append the latest prediction
    with open(filename, 'a', newline='') as file:
        writer = csv.writer(file)
        writer.writerow([timestamp, f"{actual_price:.2f}", f"{predicted_price:.2f}"])

    print(f"✅ Prediction saved to {filename}")

# Evaluate the model
def evaluate_model(y_test_inv, y_pred_inv):
    # Calculate metrics
    rmse = np.sqrt(mean_squared_error(y_test_inv, y_pred_inv))
    mae = mean_absolute_error(y_test_inv, y_pred_inv)
    r2 = r2_score(y_test_inv, y_pred_inv)

    print(f"📉 RMSE: {rmse:.2f}")
    print(f"📏 MAE: {mae:.2f}")
    print(f"📈 R² Score: {r2:.4f}")

# Main function to run the code
def main():
    # Fetch data (with delay to prevent rate limit issues)
    print("Fetching data...")
    df = fetch_data()
    if df is None:
        return

    # Preprocess data
    X, y, scaler = preprocess_data(df)
    split = int(0.8 * len(X))  # 80% train, 20% test
    X_train, X_test = X[:split], X[split:]
    y_train, y_test = y[:split], y[split:]

    # Build and train the LSTM model
    model = build_model((X.shape[1], 1))
    print("Training model...")
    model.fit(X_train, y_train, epochs=5, batch_size=32)

    # Predict
    y_pred = model.predict(X_test)

    # Inverse transform predictions and true values
    y_pred_inv = scaler.inverse_transform(y_pred)
    y_test_inv = scaler.inverse_transform(y_test.reshape(-1, 1))

    # Save prediction to CSV
    save_prediction_to_csv(y_test_inv, y_pred_inv)

    # Evaluate the model
    evaluate_model(y_test_inv, y_pred_inv)

    # Predict next Bitcoin price
    last_sequence = X[-1].reshape(1, X.shape[1], 1)
    next_pred = model.predict(last_sequence)
    print(f"Predicted Next Bitcoin Price: ${scaler.inverse_transform(next_pred)[0][0]:.2f}")

if __name__ == "__main__":
    main()
```

```python
import numpy as np
import pandas as pd
from binance.client import Client
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
import time

# Step 1: Set up the Binance client
api_key = ""  # Replace with your API key
api_secret = ""  # Replace with your API secret
client = Client(api_key, api_secret)

# Step 2: Fetch real-time Bitcoin data from Binance (1-minute intervals)
def fetch_real_time_data(symbol="BTCUSDT", interval="1m", look_back_period=100):
    # Get the last 'look_back_period' minutes of data
    klines = client.get_klines(symbol=symbol, interval=interval, limit=look_back_period)

    # Convert the data into a DataFrame
    df = pd.DataFrame(klines, columns=["timestamp", "open", "high", "low", "close", "volume", "close_time", "quote_asset_volume", "number_of_trades", "taker_buy_base_asset_volume", "taker_buy_quote_asset_volume", "ignore"])
    df["timestamp"] = pd.to_datetime(df["timestamp"], unit="ms")
    df.set_index("timestamp", inplace=True)

    # Select only the relevant columns: Open, High, Low, Close, Volume
    df = df[['open', 'high', 'low', 'close', 'volume']].astype(float)
    return df

# Step 3: Preprocess data for LSTM model
def preprocess_data(df, sequence_length=60):
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(df[['close']])  # Scale only 'close' price

    X, y = [], []
    for i in range(sequence_length, len(scaled_data)):
        X.append(scaled_data[i-sequence_length:i])
        y.append(scaled_data[i])

    X, y = np.array(X), np.array(y)
    return X, y, scaler

# Step 4: Build LSTM model
def build_model(input_shape):
    model = Sequential()
    model.add(LSTM(units=50, return_sequences=True, input_shape=input_shape))
    model.add(LSTM(units=50))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

# Step 5: Run prediction and evaluate
def main():
    # Fetch the latest real-time data from Binance
    df = fetch_real_time_data()

    # Preprocess the data for model input
    X, y, scaler = preprocess_data(df)

    # Split data (80% train, 20% test)
    split = int(0.8 * len(X))
    X_train, X_test = X[:split], X[split:]
    y_train, y_test = y[:split], y[split:]

    # Build and train the LSTM model
    model = build_model((X.shape[1], X.shape[2]))  # Update to (X.shape[1], X.shape[2]) as X has 3 dimensions now
    model.fit(X_train, y_train, epochs=5, batch_size=32)

    # Predict on the test data
    y_pred = model.predict(X_test)

    # Reshape the predictions and the true values to be 2D arrays for inverse scaling
    y_pred = y_pred.reshape(-1, 1)  # Make predictions 2D (n_samples, 1)
    y_test = y_test.reshape(-1, 1)  # Reshape y_test to match prediction shape

    # Inverse transform to get real values for Close
    y_pred_inv = scaler.inverse_transform(y_pred)  # Inverse transform only the 'close' prices
    y_test_inv = scaler.inverse_transform(y_test)

    # Calculate RMSE and other metrics
    rmse = np.sqrt(np.mean((y_test_inv - y_pred_inv) ** 2))
    print(f"RMSE: {rmse}")

    # Print the last predicted value
    print(f"Predicted Last Close Price: {y_pred_inv[-1]}")

    # Optionally: Plot the results
    import matplotlib.pyplot as plt
    plt.figure(figsize=(10, 6))
    plt.plot(y_test_inv, label="True Prices")
    plt.plot(y_pred_inv, label="Predicted Prices")
    plt.legend()
    plt.show()

if __name__ == "__main__":
    main()
```
